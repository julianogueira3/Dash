```python
import pandas as pd

dados = pd.read_csv('data/jogadores.csv')

for jogador in range(2):
    col_x = f'j{jogador}_x'
    col_y = f'j{jogador}_y'
    min_x = dados[col_x].min()
    max_x = dados[col_x].max()
    min_y = dados[col_y].min()
    max_y = dados[col_y].max()
    dados[col_x] = (dados[col_x] - min_x) / (max_x - min_x)
    dados[col_y] = (dados[col_y] - min_y) / (max_y - min_y)

dados.to_csv('jogadores.csv', index=False)

```

```python
# Função para extrair os valores x e y de uma célula da coluna no formato (x, y)
def extract_xy(cell_value):
    if isinstance(cell_value, str):
        match = re.match(r'\(([^,]+),([^)]+)\)', cell_value)
        if match:
            return float(match.group(1)), float(match.group(2))
    return None, None
```
